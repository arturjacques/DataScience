# Importação das bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

# Leitura dos Data Frames

```python
faturamento = pd.read_csv('resumo-faturamento-price-info.csv',delimiter=';')
faturamento.head()
```

```python
ocupacao = pd.read_csv('resumo-ocupacao-price-info.csv',delimiter=';')
ocupacao.head()
```

## <font color='blue'>Calculando o total de cada ID</font>

```python
meses = ['01/08/2019','01/09/2019','01/10/2019','01/11/2019','01/12/2019','01/01/2020']
soma = 0

for i in meses:
    soma = soma + faturamento[i]

faturamento['TOTAL'] = soma  
faturamento.head()
```

```python
meses = ['01/08/2019','01/09/2019','01/10/2019','01/11/2019','01/12/2019','01/01/2020']
soma = 0

for i in meses:
    soma = soma + ocupacao[i]

ocupacao['TOTAL'] = soma  
ocupacao.head()
```

## <font color='blue'>Analisando distribuição</font>

```python
plt.figure(figsize=(12,6))
distplot = sns.distplot(faturamento['TOTAL'],kde = False, axlabel='Total em R$')
plt.title('Distribuição de faturamentos')
fig = distplot.get_figure()
fig.savefig('distribuicao.png') 
```

# Qual mês apresentou o maior (e o menor) faturamento médio

```python
faturamento_medio = {}

for i in meses:
    faturamento_medio[i] =faturamento[i].mean()
    print(i,faturamento_medio[i])
```

```python
y = []
for i in meses:
    y.append(faturamento_medio[i])
```

```python
plt.figure(figsize=(12,6))
barplot = sns.barplot(x=meses,y=y,color='#add8e6')
plt.title('Faturamento médio')
plt.xlabel('Mês')
plt.ylabel('Faturamento médio em R$')
fig = barplot.get_figure()
fig.savefig('faturamento médio.png') 
```

## <font color='red'>O mês de janeiro apresentou o maior faturamento e setembro o menor faturamento médio no mês</font>

# Qual a média de taxa de ocupação para Janeiro? e para Outubro?

```python
dias_meses = {}
dias_meses['01/01/2020'] = 31
dias_meses['01/12/2019'] = 31
dias_meses['01/11/2019'] = 30
dias_meses['01/10/2019'] = 31
dias_meses['01/09/2019'] = 30
dias_meses['01/08/2019'] = 31
```

```python
y = []
for i in meses:
    y.append(ocupacao[i].mean()/dias_meses[i]) 
```

```python
plt.figure(figsize=(12,6))
barplot = sns.barplot(x=meses,y=y,color='#add8e6')
plt.title('Taxa de Ocupação')
plt.xlabel('Mês')
plt.ylabel('Taxa de ocupação')
fig = barplot.get_figure()
fig.savefig('Taxa_de_ocupacao.png') 
```

```python
ocupacao['01/01/2020'].mean()/31
```

```python
ocupacao['01/10/2019'].mean()/31
```

## <font color='red'>O mês de Janeiro teve 54% de média de ocupação enquanto o mês de outubro somente 22%</font>
